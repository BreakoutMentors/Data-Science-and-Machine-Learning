# Matplotlib Pyplot Basics

When doing any Machine Learning or Data Science work, being able to plot data is a very important skill that not only benefits you, but the people you work with. Being able to plot data is a great way to easily view trends or relationships in your data for you and other people who do not know how to code. That is the beauty of having this skill, that you are able to discuss your findings visually without having to show your code.

Matplotlib is the most popular plotting library in python, however there are others like plotly.

```
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Plotting Lines

Learning to plot lines is a great way to start with matplotlib. It might be a little intense trying to understand how everything is working, so I am going to explain the different Matplotlib methods you see below.

1. [plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html): This takes your data and different formatting options like the color of the line to customize it.
 - To have a clearer vision of how this works, it usually takes a list or numpy array for x-values and another list or array for the y-values, then it plots the (x, y) coordinates then draws a line between them.
2. [xlabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html): This is the title for the x-axis.
3. [ylabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html): This is the title for the y-axis.
4. [title()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html): This is the title to give the entire plot
5. [show()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html): This is used to display the plot that you created, which happens always happens last

The different customizations that you create for your plots don't have to happen in the order that was shown above, except `plot()` and `show()`. This adds some freedom when creating your plots.


```
# using the plt api directly
plt.plot([1, 2, 3, 4])

# add x, y axis labels and title
plt.xlabel('some numbers')
plt.ylabel('some numbers')
plt.title('Plot of some numbers')

# show the plot
plt.show();
```

```
# using the object-oriented style

x = np.linspace(0, 5, 100)

plt.plot(x, x, label='linear')  # Plot y=x
plt.plot(x, x**2, label='quadratic')  # Plot y=x^2
plt.plot(x, x**3, label='cubic')  # Plot y=x^3
plt.xlabel('x label')  # Add an x-label to the axes.
plt.ylabel('y label')  # Add a y-label to the axes.
plt.title("Simple Plot")  # Add a title to the axes.
plt.legend()  # Add a legend.

plt.show();
```

## Showing Multiple Plots

You are able to create multiple plots using subplots, I am going to separate the three lines above into their own plots in a 2x2 grid and include a plot for $y=x^4$, a 'Quartic' Polynomial.

```
# using the object-oriented style

x = np.linspace(0, 5, 100)

fig, axes = plt.subplots(2, 2)  # Create a figure and an axes.
axes[0, 0].plot(x, x, label='linear', color='blue')  # Plot some data on the axes.
axes[0, 1].plot(x, x**2, label='quadratic', color='orange')  # Plot more data on the axes...
axes[1, 0].plot(x, x**3, label='cubic', color='green')  # ... and some more.
axes[1, 1].plot(x, x**4, label='quartic', color='purple')  # ... and some more.

# Turning on legends for all
axes[0][0].legend()
axes[0][1].legend()
axes[1][0].legend()
axes[1][1].legend()


# Giving a title for the entire figure
fig.suptitle('Polynomial Graphs')

plt.show();
```

# Plotting Distributions with Histograms

To plot statistical distributions like the Normal Distribution, histograms are needed. They are plots that show the frequency of numbers in a list, array, or Pandas Series by couting the numbers that fit in a certain range called a bin. If you increase the number of bins, you can see the distributions more clearly.

This is the formula for the Probability Density Function (PDF) of a Normal Distribution:

$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}$

In the code below, it is not important to draw the line of the Probability Density Function. I did it just to show that Matplotlib's histogram is able to display these distributions. This is very useful when wanting to see the distribution of columns in some Pandas DataFrames to get a better understanding of your data.

```
# Helper Function to plot the Normal Distribution line
def normal_distribution(x, mean, std):
    """
    params:
        x (np.array): Array for all x values for input
        mean (float): Mean of the distribution
        std (float): Standard Deviation of distribution
    """
    return 1 / (np.sqrt(2*np.pi*std**2)) * np.exp((-1/2) * ((x-mean)/std)**2)
```

Below, I repeat the code twice where everything is the same except for the amount of bins. From what I explained above, the number of bins you have in your histogram is the number of groups that will have their own frequency counts. So when you increase the number of bins, the range the bins get more smaller which means it becomes more and more specific. The first histogram has 10 bins, so only 10 groups, while the second histogram has 100 bins. Please play around with the number of bins or the sample size to see how it effects the histogram.

The parameter `density` is just telling matplotlib that you are plotting a statistical distribution, so when you are plotting real data, it isn't that important.

```
std = 1.0
mean = 0.0
sample_size = 100000 # Sampling 100,000 numbers from the distribution

# Sampling from a standard normal distribution with a mean=0 and a std=1
normal_dist = np.random.normal(mean, std, sample_size)

# Getting random integers from -4 to 4 to plot line of the normal distribution
x = np.linspace(-4, 4, num=100000)

# Getting the y's to plot the line for the normal distribution
y = normal_distribution(x, mean, std)
# The density parameter means that you are telling it that this is a probability density function
plt.hist(normal_dist, bins=10, density=True)

plt.plot(x, y, color='r', label='Normal Distribution PDF')
plt.legend()
plt.show();
```

```
std = 1.0
mean = 0.0
sample_size = 100000 # Sampling 100,000 numbers from the distribution

# Sampling from a standard normal distribution with a mean=0 and a std=1
normal_dist = np.random.normal(mean, std, sample_size)

# Getting random integers from -4 to 4 to plot line of the normal distribution
x = np.linspace(-4, 4, num=100000)

# Getting the y's to plot the line for the normal distribution
y = normal_distribution(x, mean, std)

# The density parameter means that you are telling it that this is a probability density function
plt.hist(normal_dist, bins=100, density=True)

plt.plot(x, y, color='r', label='Normal Distribution PDF')
plt.legend()
plt.show();
```

## Scatter Plots: Viewing relationships between variables

Scatter Plots are plots that show the relationship of two different variables, like two Pandas columns in a DataFrame. The reason you would want to do this is visualize if they have a "linear" relationship in a positive or negative direction.

I will use a dataset that you will see in our Linear Regression lessons later, that describes life expectancy of different countries along with other health and economic data.

```
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/BreakoutMentors/Data-Science-and-Machine-Learning/main/datasets/Life%20Expectancy%20Data.csv")
df.columns = [col_name.strip() for col_name in df.columns] # used to fix weird spaces in the column names
df.head() # view the first 5 rows of the data
```

If I want to see the relationship between `Schooling` and `Life expectancy` columns, a scatter plot will do that. Schooling details the average amount of years that a citizen undergoes education in the country, while Life expectancy details the average life-span of a person in that country.

```
plt.scatter(x=df['Schooling'], y=df['Life expectancy'])
plt.xlabel('Schooling')
plt.ylabel('Life expectancy')
plt.show()
```

From this plot, you are able to see that there is a trend between these two variables. Especially that as the number of years of schooling increases, the life expectancy of that country increases as well. So it suggests that education can have a positive impact on the overall well-being of citizens globally.

And you might see that this looks really familiar to the scatter plot function you saw in the Pandas lesson. Below I will use the pandas DataFrame `df` to plot the scatter plot.

```
df.plot.scatter(x='Schooling', y='Life expectancy');
```

You can see that they are VERY similar with some formatting changes. The reason they look alike is because they both use Matplotlib, the developers of Pandas use matplotlib in the background when using their plotting features. So they both get the job done, so please feel free to choose.
