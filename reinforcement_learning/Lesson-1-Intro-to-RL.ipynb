# Policy and Value-Based Methods
---

**Policy-Based Method**:
  - Policy is the agent's **BRAIN** !
    - In the game above, the policy will *guide* me what action to take when I am on different coordinate!
      - So technically, given a *state* (where I am at), a policy will output an *action*!

**Value-Based Method**:
  - This method is actually the policy-based method in *disguise*
    - Instead of finding the best *policy*, we are trying to find the best *value function*
      - A **value function** is a machine that accepts a state (usually) as input and ouputs a value
        - The *magnitude* of that value will tell us which action is better than other!
